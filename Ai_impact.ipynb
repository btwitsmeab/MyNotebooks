```python
dfClean = df.copy()
dfClean.head()
```

```python
# ordinal / binary
year_map = {
    'Freshman': 1,
    'Sophomore': 2,
    'Junior': 3,
    'Senior': 4
}

dfClean['Year_of_Study'] = dfClean['Year_of_Study'].map(year_map)

dfClean['Prompt_Engineering_Skill'] = dfClean['Prompt_Engineering_Skill'].map({
    'Beginner': 0,
    'Intermediate': 1,
    'Advanced': 2
})

dfClean['Paid_Subscription'] = dfClean['Paid_Subscription'].astype(int)

dfClean['Burnout_Risk_Level'] = dfClean['Burnout_Risk_Level'].map({
    'Low': 0,
    'Medium': 1,
    'High': 2
})
```

```python
onehot_cols = [
    'Major_Category',
    'Primary_Use_Case',
    'Institutional_Policy'
]

dfClean = pd.get_dummies(dfClean, columns=onehot_cols, drop_first=True)
```

```python
dfClean.info()
```

```python
dfClean['GPA_Improvement'] = (
    dfClean['Post_Semester_GPA']
    - dfClean['Pre_Semester_GPA']
)

target = 'GPA_Improvement'

features = [
    'Weekly_GenAI_Hours',
    'Tool_Diversity',
    'Paid_Subscription',
    'Perceived_AI_Dependency',
    'Traditional_Study_Hours',
    'Anxiety_Level_During_Exams'
]
X = dfClean[features]
y = dfClean[target]
```

```python
from sklearn.model_selection import KFold, StratifiedKFold, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold

kf = KFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error

r2_scores = []
mae_scores = []

for train_idx, test_idx in kf.split(X):

    X_train = X.iloc[train_idx]
    X_test = X.iloc[test_idx]

    y_train = y.iloc[train_idx]
    y_test = y.iloc[test_idx]

    model = RandomForestRegressor(random_state=42)

    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    r2_scores.append(r2_score(y_test, y_pred))
    mae_scores.append(mean_absolute_error(y_test, y_pred))

print("Mean R²:", np.mean(r2_scores))
print("Mean MAE:", np.mean(mae_scores))
```
